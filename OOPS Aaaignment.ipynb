```python
### Q.3  Explain the difference between instance methods and class methods. Provide an example of each.

ans)    ## In Python, instance methods and class methods are two types of methods that can be defined in a class.

## Instance Methods:

# - These methods are bound to an instance of a class.
# - They have access to the instance's attributes and methods.
# - They are called on an instance of the class.
# - The first argument is always the instance itself, referred to as self.

## Example:


class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        print(f"{self.name} says Woof!")

my_dog = Dog("Fido")
my_dog.bark()  # Output: Fido says Woof!


## Class Methods:

# - These methods are bound to the class itself, not an instance.
# - They have access to the class's attributes and methods.
# - They are called on the class, not an instance.
# - The first argument is always the class itself, referred to as cls.

## Example:


class Dog:
    num_dogs = 0

    def __init__(self, name):
        self.name = name
        Dog.num_dogs += 1

    @classmethod
    def get_num_dogs(cls):
        return cls.num_dogs

print(Dog.get_num_dogs())  # Output: 0
my_dog = Dog("Fido")
print(Dog.get_num_dogs())  # Output: 1


## In summary, instance methods operate on instances of a class, while class methods operate on the class itself.
```

```python
### Q.4 How does Python implement method overloading? Give an example.

ans)    ## Python does not support method overloading in the same way that languages like Java or C++ do. In Python, you can't define multiple methods with the same name but different parameters.

## However, Python offers alternative solutions to achieve similar behavior:

# 1. Default Argument Values: You can define a method with default argument values, allowing for varying numbers of arguments.

Example:


def greet(name, msg="Hello"):
    print(f"{msg}, {name}")

greet("John")  # Output: Hello, John
greet("John", "Hi")  # Output: Hi, John


# 1. Variable-length Argument Lists: You can use *args or **kwargs to accept varying numbers of arguments.

Example:


def sum_numbers(*args):
    return sum(args)

print(sum_numbers(1, 2, 3, 4))  # Output: 10


# 1. Single Dispatch: You can use the @singledispatch decorator from the functools module to implement a form of method overloading.

Example:


from functools import singledispatch

@singledispatch
def fun(arg):
    return "Default"

@fun.register
def _(arg: int):
    return "Integer"

@fun.register
def _(arg: list):
    return "List"

print(fun(1))  # Output: Integer
print(fun([1, 2]))  # Output: List
print(fun("a"))  # Output: Default


##These approaches allow for flexible method definitions, achieving similar results to method overloading in other languages.
```

```python
### Q.6 Describe the five types of inheritance in Python. Provide a simple example of multiple inheritance.

ans)     ## Python supports five types of inheritance:

# 1. Single Inheritance: A child class inherits from a single parent class.
# 2. Multiple Inheritance: A child class inherits from multiple parent classes.
# 3. Multilevel Inheritance: A child class inherits from a parent class, which itself inherits from another parent class.
# 4. Hierarchical Inheritance: Multiple child classes inherit from a single parent class.
# 5. Hybrid Inheritance: A combination of multiple and multilevel inheritance.

## Here's an example of multiple inheritance:


class Animal:
    def eat(self):
        print("Eating...")

class Mammal:
    def walk(self):
        print("Walking...")

class Dog(Animal, Mammal):
    def bark(self):
        print("Barking...")

my_dog = Dog()
my_dog.eat()  # Output: Eating...
my_dog.walk()  # Output: Walking...
my_dog.bark()  # Output: Barking...


## In this example, the Dog class inherits methods from both the Animal and Mammal classes, demonstrating multiple inheritance.
```

```python
### Q.7  What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?

ans)     ## The Method Resolution Order (MRO) in Python is the order in which a class searches for methods and attributes in its inheritance graph. It's used to resolve conflicts when multiple parent classes have methods with the same name.

## To retrieve the MRO programmatically, you can use the mro() method or the __mro__ attribute:

# 1. mro() method:


class Animal:
    pass

class Mammal(Animal):
    pass

class Dog(Mammal):
    pass

print(Dog.mro())  # Output: [<class '__main__.Dog'>, <class '__main__.Mammal'>, <class '__main__.Animal'>, <class 'object'>]


# 2. __mro__ attribute:


print(Dog.__mro__)  # Output: (__main__.Dog, __main__.Mammal, __main__.Animal, object)


## The MRO is calculated using the C3 linearization algorithm, which ensures that:

# - A class is searched before its parents.
# - Parents are searched in the order they're listed in the class definition.
# - If a parent is listed multiple times, it's only searched once.

```

```python
### Q.8  Create an abstract base class 'Shape` with an abstract method area(). Then create two subclasses Circle and Rectangle that implement the 'area() method.

ans)    ## Here's an example implementation:


from abc import ABC, abstractmethod
import math

# Abstract base class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Subclass 1: Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

# Subclass 2: Rectangle
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Create instances and call the area() method
circle = Circle(5)
print(f"Circle area: {circle.area()}")

rectangle = Rectangle(4, 6)
print(f"Rectangle area: {rectangle.area()}")


## In this example:

# - The Shape class is an abstract base class with an abstract area() method.
# - The Circle and Rectangle classes inherit from Shape and implement the area() method.
# - We create instances of Circle and Rectangle and call their area() methods to calculate their respective areas.

## This demonstrates polymorphism, as the area() method behaves differently depending on the subclass instance.
```

```python
### Q.9   Demonstrate polymorphism by creating a function that can work with different shape objects to calculate and print their areas.

ans)  ##  Here's an example:


from abc import ABC, abstractmethod
import math

# Abstract base class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Subclass 1: Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

# Subclass 2: Rectangle
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Polymorphic function
def print_area(shape: Shape):
    print(f"Area: {shape.area()}")

# Create instances and call the polymorphic function
circle = Circle(5)
print_area(circle)  # Output: Area: 78.53981633974483

rectangle = Rectangle(4, 6)
print_area(rectangle)  # Output: Area: 24


## In this example:

#- The print_area() function takes a Shape object as an argument.
# - It calls the area() method on the passed object, without knowing its specific subclass.
# - The correct area() method is called based on the object's class (Circle or Rectangle), demonstrating polymorphism.

## This allows us to write a single function that works with different shape objects, making our code more flexible and reusable.
```

```python
### Q.10 Implement encapsulation in a BankAccount class with private attributes for 'balance and account_number. Include methods for deposit, withdrawal, and balance inquiry.

ans)   ## Here's an example implementation:


class BankAccount:
    def __init__(self, account_number, initial_balance):
        self.__account_number = account_number
        self.__balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.__balance}")
        else:
            print("Invalid withdrawal amount.")

    def check_balance(self):
        print(f"Account balance: ${self.__balance}")

    def get_account_number(self):
        return self.__account_number

# Example usage:
account = BankAccount("1234567890", 1000)
print(f"Account Number: {account.get_account_number()}")
account.check_balance()
account.deposit(500)
account.withdraw(200)
account.check_balance()


## In this example:

# - The BankAccount class encapsulates the balance and account_number attributes by making them private (__balance and __account_number).
# - The class provides methods for deposit, withdrawal, and balance inquiry, which modify or access the private attributes.
# - The get_account_number method allows access to the account number, while keeping it read-only.
# - The example usage demonstrates creating an account, checking the balance, depositing and withdrawing funds, and checking the balance again.

## This implementation ensures that the sensitive account information is protected from direct access and modification, promoting data security and integrity.
```

```python
### Q.12  Create a decorator that measures and prints the execution time of a function.

ans)    ## Here's an example of a decorator that measures and prints the execution time of a function:


import time
from functools import wraps

def measure_execution_time(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"Function {func.__name__} executed in {execution_time:.4f} seconds")
        return result
    return wrapper

# Example usage:
@measure_execution_time
def example_function():
    time.sleep(1)  # Simulate some work
    print("Function executed")

example_function()


## In this example:

# - The measure_execution_time decorator uses the time module to record the start and end times of the function execution.
# - The wraps decorator from the functools module is used to preserve the original function's metadata (name, docstring, etc.).
# - The decorator prints the execution time of the function after it finishes executing.

## When you run example_function, it will print:


#Function executed
#Function example_function executed in 1.0001 seconds


## This decorator can be applied to any function to measure and print its execution time.
```

```python
### Q.13   Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?

ans)     ## The Diamond Problem occurs in multiple inheritance when two classes inherit from a common base class, and a fourth class inherits from both of them. This creates a diamond shape in the inheritance graph.


   A
  / \
 B   C
  \ /
   D


## In this scenario, if classes B and C override a method from class A, and class D inherits from both B and C, it's unclear which overridden method D should use.

## Python resolves the Diamond Problem using:

# 1. Method Resolution Order (MRO): Python calculates the MRO for each class, which is the order in which classes are searched for methods.
# 2. C3 Linearization: Python uses the C3 linearization algorithm to compute the MRO. This algorithm ensures that:
    # - A class is searched before its parents.
    # - Parents are searched in the order they're listed in the class definition.
    # - If a parent is listed multiple times, it's only searched once.

## In the Diamond Problem scenario, Python's MRO would be:


D, B, C, A, object


## This means that Python will first search for methods in class D, then in class B, then in class C, and finally in class A. If classes B and C override a method from class A, class D will use the overridden method from the first class it encounters in the MRO, which is class B.

## Python's approach to the Diamond Problem ensures that:

# - Methods are resolved consistently and predictably.
# - The order of inheritance matters, but the algorithm ensures that the most specific implementation is used.

## By understanding the Diamond Problem and Python's resolution strategy, you can design and use multiple inheritance effectively in your Python programs.
```
